```python
movies_with_genres = df.copy(deep=True)

x = []
for index, row in df.iterrows():
    x.append(index)
    for genre in row['listed_in']:
        movies_with_genres.at[index, genre] = 1
    for type1 in row['type']:
        movies_with_genres.at[index, type1] = 1
    for duration in row['duration']:
        movies_with_genres.at[index, duration] = 1

print(len(x) == len(df))

movies_with_genres.head(3)
```

```python
movies_with_genres = movies_with_genres.fillna(0)
movies_with_genres.head(3)
```

```python
# First, let's reset index to default and drop the existing index.
movies_with_genres.reset_index(drop=True, inplace=True)

# Next, let's drop redundant columns
movies_with_genres.drop(['show_id','title','type','duration', 'listed_in', 'release_year'], axis=1, inplace=True)

# Let's view chamges

movies_with_genres
```
