- Transformando as colunas 'listed_in & director' em listas separadas por ','

```python
df['listed_in'] = df.listed_in.str.split(',')
```

```python
df['director'] = df.director.str.split(',')
```

```python
df.head(3)
```

```python
# Pegando somentes as colunas relevantes
df = df[['show_id', 'title', 'type', 'duration','listed_in', 'release_year']]
```

```python
# vendo o tanto de memoria alocada em para cada coluna
movies_df_original_mem = df.memory_usage()
movies_df_original_mem
```

```python
df.info()
```

```python
# alterando o tipo de coluna para locar menos memoria
df.show_id = df.show_id.astype('int32')
```

```python
df.isna().sum()
```

```python
df.release_year = df.release_year.astype('int16')
```

```python
movies_df_new_mem = df.memory_usage()

print(movies_df_original_mem)
print()
print(movies_df_new_mem)
```

```python
df.head(3)
```

```python
# transformando todos os tipos de caterogias da coluna
# 'listed_in' em colunas. Para criar um modelo capaz de detectar gosto
# por meio do gosto de categoria, onde o valor 1 diz q tem aquela categoria
# e o valor 0 não tem.
movies_listed_in = df.copy(deep=True)

x = []
for index, row in df.iterrows():
    x.append(index)
    for genre in row['listed_in']:
        movies_listed_in.at[index, genre] = 1

print(len(x) == len(df))

movies_listed_in.head(3)
```

```python
movies_listed_in = movies_listed_in.fillna(0)
movies_listed_in.tail()
```

```python
X_cat = movies_listed_in[['type', 'duration']]
```

```python
# fazendo a msm coisa para as colunas type e duration
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
X_cat = pd.DataFrame(X_cat)
X_cat
```

```python
# juntando todas as colunas em unico dataset
movies_listed_in = pd.concat([movies_listed_in, X_cat], axis=1, join='inner')
```

```python
movies_listed_in.head()
```

```python
# gerando gosto de uma "pessoa" referente aos filme/series 
# do dataset, onde rating é a avaliaçao do filme
movie_ratings = [
            {'title':'Friends', 'rating':4.9},
            {'title':'Maron', 'rating':4.9},
            {'title':'Red vs. Blue', 'rating':4.7},
            {'title':'Transformers Prime', 'rating':3}
         ] 
movie_ratings = pd.DataFrame(movie_ratings)
movie_ratings
```

```python
# Pegando os filmes no dataset e adicionando a coluna rating a ela
movie_title = df[df['title'].isin(movie_ratings['title'])]
movie_ratings = pd.merge(movie_title, movie_ratings)
movie_ratings
```

```python
movies_listed_in_df = movies_listed_in[movies_listed_in.show_id.isin(movie_ratings.show_id)]
movies_listed_in_df
```

```python
# já que só precisamos das colunas referente aos gostos em categoria, tipo e duranção, só valos utilizar essas colunas.
movies_listed_in_df.reset_index(drop=True, inplace=True)

movies_listed_in_df.drop(['show_id','title','type','duration', 'listed_in', 'release_year'], axis=1, inplace=True)
movies_listed_in_df
```

```python
# priorizando o gosto com base na avaliação da "pessoa"
movie_profile = movies_listed_in_df.T.dot(movie_ratings.rating)

movie_profile
```

```python
movies_listed_in = movies_listed_in.set_index(movies_listed_in.show_id)

movies_listed_in.head()
```

```python
movies_listed_in.drop(['show_id','title', 'type','duration','listed_in', 'release_year'], axis=1, inplace=True)

movies_listed_in.head()
```

```python
# Analisando gosto por media poderada
```

```python
recommendation_table_df = (movies_listed_in.dot(movie_profile)) / movie_profile.sum()

recommendation_table_df.head()
```

```python
recommendation_table_df.sort_values(ascending=False, inplace=True)

recommendation_table_df.head(20)
```

```python
copy = df.copy(deep=True)
copy = copy.set_index('show_id', drop=True)

top_20_index = recommendation_table_df.index[:20].tolist()

recommended_movies = copy.loc[top_20_index, :]
recommended_movies
```
